```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from text file
data = np.loadtxt(r'F:\Research\Drought Thesis\Analysis\SDSM\Temperature Processed data\Monthly Converted\Bangladesh Tmax Historical.csv', delimiter=',')

# Reshape data into a 2D array with one row per year
data = data.reshape(-1, 12)

# Create an array of months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create an array of angles
angles = np.linspace(0, 2 * np.pi, len(months), endpoint=False)

# Create polar plot
ax = plt.subplot(111, projection='polar')

# Plot data for each year
for year_data in data:
    ax.plot(angles, year_data)

# Set xticks and labels
ax.set_xticks(angles)
ax.set_xticklabels(months)

# Set radial axis limits
ax.set_ylim(28, 30)

# Set title
ax.set_title('Polar Diagram from Historical Rainfall (mm)')

# Show plot
plt.show()
```

```python
import pandas as pd
import os

file_path = r'F:\Research\Drought Thesis\Analysis\SDSM\Temperature Processed data\Tmin_historical Downscaled.csv'

df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
df.set_index('Date', inplace=True)

# Calculate the average of all stations for each month
monthly_avg = df.groupby(pd.Grouper(freq='M')).mean().mean(axis=1)

# Save the result to a CSV file
output_folder = r'F:\Research\Drought Thesis\Analysis\SDSM\Temperature Processed data\Monthly Converted'
file_name = 'Bangladesh Tmin Historical.csv'
monthly_avg.to_frame()[0].to_csv(os.path.join(output_folder, file_name), header=False, index=False)

print(monthly_avg)
```
