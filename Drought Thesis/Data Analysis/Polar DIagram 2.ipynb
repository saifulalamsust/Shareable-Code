```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = 'F:\\Research\\Drought Thesis\\Analysis\\SDSM\\Calibration And Validation MPI\\Comparison Table\\Tmax_Observed.xlsx'

# Read the data from the excel file
df = pd.read_excel(file_path, index_col=0)

# Convert the month names to numerical values
df['Month'] = pd.to_datetime(df.index, format='%B').month

# Set the figure size
plt.figure(figsize=(8, 8))

# Create the polar plot
ax = plt.subplot(111, projection='polar')
ax.plot(df['Month'], df['Tmax'], 'ro', linewidth=2)

# Set the title and labels
ax.set_title('Polar Diagram of Tmax')
ax.set_rlabel_position(0)
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Show the plot
plt.show()
```
