```python
import pandas as pd
import os

# Read the excel file
df = pd.read_excel(r'F:\Research\Drought Thesis\Analysis\SDSM\Climate Obs_1981_2010_Processed.xlsx', sheet_name='Tmax_Filled')


df.head()
```

```python
# Get the station names from the first row of the dataframe
station_names = df.columns[0:].values.tolist()

print(station_names)
```

```python
# Specify the output folder
output_folder = r'F:\Research\Drought Thesis\Analysis\SDSM\Temperature Processed data\T_Max'

# Save data for each station as a separate DAT file
for station_name in station_names:
    station_data = df[station_name].copy()
    output_file_name = f'{station_name}_Tmax_OBS.DAT'
    output_file_path = os.path.join(output_folder, output_file_name)
    station_data.to_csv(output_file_path, sep='\t', header=False, index=False)

print("Data saved successfully!")
```
